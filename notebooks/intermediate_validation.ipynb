# Checking for Negative `length_of_stay` values

```python
import duckdb
import pandas as pd

con = duckdb.connect("../data/shelterflow.duckdb")
stays = con.execute("SELECT * FROM int_animal_stays").df()

pd.set_option("display.max_rows", None)
```

```python
negatives = stays[stays["length_of_stay"] < 0]
print(len(negatives))
print(negatives["length_of_stay"].describe())
negatives.sort_values("length_of_stay").head(20)
```
