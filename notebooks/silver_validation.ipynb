## Investigating Intake/Outcome Count Mismatch

```python
# Count intakes per animal
intake_counts = silver_intakes.groupby("animal_id").size().rename("n_intakes")

# Count outcomes per animal
outcome_counts = silver_outcomes.groupby("animal_id").size().rename("n_outcomes")

# Combine into one table
counts = pd.concat([intake_counts, outcome_counts], axis=1).fillna(0)

# How many animals have mismatched counts?
mismatched = counts[counts["n_intakes"] != counts["n_outcomes"]]
print(f"Animals with mismatched intake/outcome counts: {len(mismatched)}")
print(f"Total animals: {len(counts)}")
print(f"Percentage mismatched: {len(mismatched) / len(counts) * 100:.1f}%")
```

```python
# See the breakdown of how counts differ
counts["difference"] = counts["n_intakes"] - counts["n_outcomes"]
print(counts["difference"].value_counts().sort_index())
```

```python
clean = counts[(counts["n_intakes"] == 1) & (counts["n_outcomes"] == 1)]
print(f"Animals with exactly 1 intake and 1 outcome: {len(clean)} ({len(clean)/len(counts)*100:.1f}%)")
```

71,726 animals (98%) have matching intake and outcome counts and pair directly by chronological sequence. 522 animals have one unmatched intake (likely still in shelter at the dataset's end), and 1,021 have one unmatched outcome (likely intakes predating the dataset's 2013 start). All mismatches are off-by-one; unmatched events are excluded from stay-level metrics via an inner join, an assumption appropriate for aggregate analysis.
