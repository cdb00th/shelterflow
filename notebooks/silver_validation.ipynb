```python
import duckdb
import pandas as pd

con = duckdb.connect("../data/shelterflow.duckdb")
bronze_intakes = con.execute("SELECT * FROM bronze_intakes").df()
bronze_outcomes = con.execute("SELECT * FROM bronze_outcomes").df()
silver_intakes = con.execute("SELECT * FROM silver_intakes").df()
silver_outcomes = con.execute("SELECT * FROM silver_outcomes").df()

pd.set_option("display.max_rows", None)
```

## Verifying `age_group` populates sensibly

```python
silver_intakes[["age_upon_intake", "age_group"]].value_counts().reset_index(name="count")
```
